```python
#TOP 10 CHARACTERS IN DC IN TERMS APPEARANCE
import matplotlib.pyplot as plt

sns.set_style("darkgrid")


att = dc.sort_values('APPEARANCES', ascending = False).head(10)

plt.figure(figsize=(12,7))
sns.barplot(att['name'],att['APPEARANCES'],linewidth=1,edgecolor="k"*len(att))
plt.xticks(rotation='vertical')
plt.grid(True)
plt.title("TOP 10 DC CHARACTERS APPEARANCE",color='b')
plt.show()
```

```python
#TOP 10 CHARACTERS IN MARVEL IN TERMS APPEARANCE

sns.set_style("darkgrid")


att1 = marvel.sort_values('APPEARANCES', ascending = False).head(10)

plt.figure(figsize=(12,7))
sns.barplot(att1['name'],att1['APPEARANCES'],linewidth=1,edgecolor="k"*len(att))
plt.xticks(rotation='vertical')
plt.grid(True)
plt.title("TOP 10 MARVEL CHARACTERS APPEARANCE",color='b')
plt.show()
```
